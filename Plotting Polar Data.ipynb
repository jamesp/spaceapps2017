```python
import os
import time
```

```python
files = [os.path.join('monthly', f) for f in os.listdir('./monthly/')]
```

```python
ds = xr.open_mfdataset(files)
```

```python
fig, ax = plt.subplots()
for t in ds.time:
    ax.clear()
    ds.seaice_conc_monthly_cdr.sel(time=t).plot.contourf(add_colorbar=False, ax=ax)
    fig.canvas.draw()
    time.sleep(1)
```
