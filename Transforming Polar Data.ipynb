```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import xarray as xr

import cartopy.crs as ccrs
import cartopy

import os
import time
```

```python
directory = 'north/monthly/'
files = sorted([os.path.join(directory, f) for f in os.listdir(directory)])
```

```python
ds = xr.open_mfdataset(files)
```

```python
ds.seaice_conc_monthly_cdr.to_netcdf('north.nc')  # save the combined datafile
```

```python
max_month = ds.seaice_conc_monthly_cdr.sum(('xgrid', 'ygrid')).argmax('time').load()
```

```python
nps = ccrs.NorthPolarStereo(central_longitude=-45)
sps = ccrs.SouthPolarStereo(central_longitude=-45)
```

```python
t = ds.time[0]
fig = plt.figure(figsize=(12, 12))
ax = plt.axes(projection=ccrs.Orthographic(central_longitude=0, central_latitude=75))

pp = ds.sel(time=t).seaice_conc_monthly_cdr.plot.contourf(ax=ax, transform=nps, cmap=plt.cm.Blues_r, add_colorbar=False)
ax.set_global()

ax.add_feature(cartopy.feature.LAND)
ax.coastlines()

#(ds.seaice_conc_monthly_cdr.isel(time=28) > 0.1).plot.contour(colors='red', levels=1, transform=nps, ax=ax)
#plt.savefig(time.strftime("%Y.%m.%d.png", time.gmtime(t.astype(int)/1000000000)))
plt.title(time.strftime("%Y.%m.%d.png", time.gmtime(t.astype(int)/1000000000)))
```

```python
ds.seaice_conc_monthly_cdr.sum(('xgrid', 'ygrid')).groupby('time.year')
```

```python
i = 0
for t in ds.time:
    if time.strftime("%m", time.gmtime(t.astype(int)/1000000000)) == '03':
        i = i+1
        fig = plt.figure(figsize=(12, 12))
        ax = plt.axes(projection=ccrs.Orthographic(central_longitude=0, central_latitude=75))

        pp = ds.sel(time=t).seaice_conc_monthly_cdr.plot.contourf(ax=ax, transform=nps, cmap=plt.cm.Blues_r, add_colorbar=False)
        ax.set_global()

        ax.add_feature(cartopy.feature.LAND)
        ax.coastlines()

        (ds.seaice_conc_monthly_cdr.isel(time=28) > 0.1).plot.contour(colors='red', levels=1, transform=nps, ax=ax)
        #plt.savefig(time.strftime("%Y.%m.%d.png", time.gmtime(t.astype(int)/1000000000)))
        plt.title(time.strftime("%Y.%m.%d.png", time.gmtime(t.astype(int)/1000000000)))
        plt.savefig('annual%d.png' % i)
        print('done %r' % time.strftime("%Y.%m.%d.png", time.gmtime(t.astype(int)/1000000000)))
        plt.close('all')
```
