```python
import numpy
from matplotlib.colors import rgb2hex
import matplotlib.pyplot as plt
import geojson
from geojson import Feature, LineString, FeatureCollection


for t in ds.time:
    timestamp = time.strftime("%Y.%m.%d", time.gmtime(t.astype(int)/1000000000))
    figure = plt.figure()
    ax = figure.add_subplot(111)

    contour = (ds.seaice_conc_monthly_cdr.sel(time=t) > 0.2).plot.contour(ax=ax, levels=1)


    line_features = []
    for collection in contour.collections:
        paths = collection.get_paths()
        for path in paths:
            v = path.vertices
            coordinates = []
            for i in range(len(v)):
                lat, lon = g.transform_point(v[i][0], v[i][1], nps)
    #             lat = v[i][0]
    #             lon = v[i][1]
                coordinates.append((lat, lon))
            line = LineString(coordinates)
            properties = {
                "stroke-width": 3,
                "time":  timestamp,
            }
            line_features.append(Feature(geometry=line, properties=properties))

    feature_collection = FeatureCollection(line_features)
    geojson_dump = geojson.dumps(feature_collection, sort_keys=True)
    with open('geoout/%s.geojson' % timestamp, 'w') as fileout:
        fileout.write(geojson_dump)
    plt.close('all')
    print(timestamp)
```
